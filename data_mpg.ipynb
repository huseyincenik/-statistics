```python
# import matplotlib.pyplot as plt

# bar and pie charts are used for categorical data.

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df.origin.value_counts().index
sizes = df.origin.value_counts().values
explode = ( 0, 0.2, 0)  # only "explode" the 2nd slice (i.e. 'Hogs') # explode ; place to explode

fig, ax = plt.subplots(figsize=(7,7))

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=110) # scratching operation
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

* Skewness and kurtosis  

```python
from scipy.stats import kurtosis, skew
```

```python
# This line of code generates 10 million random numbers with a normal (Gaussian) distribution with a mean of 0 
# and a standard deviation of 2 using the NumPy library.

# Here, the first parameter of the np.random.normal function is the mean value and the second parameter is the standard deviation.
# The third parameter is the number of random numbers to be generated.
# This line of code can be used for random number generation, which can be used, for example, in a simulation or data analysis.

x=np.random.normal(0,2,10000000)
plt.hist(x,bins=100);

# The first parameter of the plt.hist() function will be the array or list containing the data of the histogram to be created, ie x. 
# The second parameter determines how many parts (thousands) the histogram will be divided into and is given here as bins=100.
```

```python
kurtosis(x) # kurtosis
```

```python
skew(x) # skewness
```

```python
# gamma curve

shape, scale = 2, 2
s=np.random.gamma(shape,scale, 100000)
```

```python
plt.hist(s, bins=100); 
```

```python
kurtosis(s) 
# Kurtosis refers to the measure of the peakness of a data set and gives information about the skewness of the distribution.
#  The kurtosis value is measured relative to 0, which is the spike of the normal distribution.

# If a data set is symmetric with respect to the normal distribution, the kurtosis value approaches 0. 
# A positive kurtosis value indicates a more pointed distribution than the normal distribution, while a negative kurtosis value indicates a more even (flattened) distribution.
```

```python
skew(s) # positive because it is skewed to the right
```
